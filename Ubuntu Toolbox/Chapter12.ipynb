## Running Commands to Browse the Web
- text-mode web broswers can be used to check if web server is working (*lynx*)
- need to install *lynx, links, elinks* (elinks for a cammand line web browser)
- *elinks* can handle most basic HTML content and features, but no images
- *elinks* supports multiple colors
> 
> - ESC: toggle the menu on and off
> - Down arrow: go to the next link or editable field on the page
> - up arrow: go to the previous link or editable field on the page
> - right arrow or enter: go forward to the highlighted link. Enter text in the highlighted form field
> - left arrow: go back to the previous page
> - /: search forward
> - ?: search backwards
> - n: find next
> - N: find previous
> - q or Ctrl+c: exit elinks
> - a: bookmark the current page
> - t: open a new browser tab


## Transferring Files
- *wget, ftp, lftp, scp, sftp*: commands for downloading files from remote servers
### Downloading Files with wget
- download files from web servers and FTP servers
- enter login and password if FTP server requires
- download a single web page
> 
- -p: download web page and other elements
- -pk: download pages, use local file names
- -pkK: rename to loca names, keep original
- -m: adds timestamping and keeps FTP directory listings


### Transferring Files with cURL
- similar to wget
- can transfer files using other popular protocols including SSH protocols
- designed for single-shot file transfers


### Transferring Files with FTP Commands
- *lftp* usage recommended
- FTP commands are similar to shell commandslft
> 
- commands preceded by an exclamation mark is executed by the local shell


### Using SSH Tools to Transfer Files
- *scp* command allows to get a file from one computer to another while making the communication safe
- *scp*: host-to-host file copy command


### Copying Remote Files with scp
- to use scp, SSH service must be running on the remote system
- -p: preserve permissions and timestamps
- -P: indicate port
- -r: recursive copy


### Copying Files Using rsync
- lets you copy files between remote systems over the *ssh* facility\
- powerful when doing backups or syncing directories over a network
- *rsync* only copies files that have changed
- --links: symbolic link files are copied (may not be enough)
- --copy-links: files that the symbolic link ultimately points to is copied to the remote location (may be too much)


### Copying Remote Files in sftp and lftp Sessions
- *sftp*: lets you to find and copy files over SSH protocols
> **sftp chris@server1**
- use in the same manner as regular FTP clients
- *lftp*: path completion features
> **lftp sftp://chris@server1**

## Sharing Remote Directories
### Sharing Remote Directories with NFS
- *exportfs* and *showmount* to see available and mounted shared directories
#### Viewing and Exporting NFS Shares
- exportfs command not found
#### Mounting NFS Shares
- sudo mkdir /mnt/server-share
- sudo mount server.example.com:/export/myshare /mnt/server-share
- sudo mount -o rw,hard,intr server.example.com:/export/myshare /mnt/server-share
- sudo mount -t nfs4 server.example.com:/ /mnt/server-share

### Sharing Remote Directories with Samba
- open source implementation of the Windows file and print sharing protocol called Common Internet File System(CIFS)
- SWAT
#### Viewing and Accessing Samba Shares
- scan network for SMB hosts
> 
> "there is none"
- view a text representation of network neighborhood
> **sudo smbtree**
- add an existing Linux user as a Samba user
> **sudo smbpasswd -a chris**
- list services offered by a server to an anonymous user
> **smbclient -L server**
- connect to a Samba share FTP-style
> **smbclient //thompson/homes -U chris**
#### Mounting Samba Shares
- to mount remote Samba shares on local filesystem:
> **sudo mount -t cifs -o username=chris, password=MySecret //192.168.1.1/homes /mnt/mymount/
- to see the current connections and file locks
> **sudo smbstatus**
#### Looking Up Samba Hosts
- determine the IP address of a computer
> **nmblookup thompson**
- find the IP address for a server on a specific subnet (-U option)
> **nmblookup -U 192.168.1.255 einstein

- check main Samba configuration file

### Sharing Remote Directories with SSHFS
- steps for mounting a directory of documents from a remote server to a local directory
1. **sudo apt-get install sshfs**
2. **mkdir /var/tmp/chris**
3. **sshfs chris@10.0.0.5:/home/chris /var/tmp/chris**
- when done, unmount: **fusermount -u /var/tmp/chris**
